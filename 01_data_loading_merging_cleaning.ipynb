```python
import zipfile
import os

zip_files = ['archive (7).zip', 'archive (8).zip', 'archive (9).zip']

for zip_file in zip_files:
    print(f"Extracting {zip_file}...")
    with zipfile.ZipFile(zip_file, 'r') as z:
        z.extractall('.')
    print(f"Done!")

print("\nAll files extracted!")
print(os.listdir('.'))
```

```python
import pandas as pd


# List all yearly files
files = [
    "2014_Financial_Data.csv",
    "2015_Financial_Data.csv",
    "2016_Financial_Data.csv",
    "2017_Financial_Data.csv",
    "2018_Financial_Data.csv"
]

# Store all dataframes
df_list = []
```

```python

for file in files:
    df = pd.read_csv(file)
    year = file.split("_")[0]
    df['year'] = int(year)
    
    df_list.append(df)

financials = pd.concat(df_list, ignore_index=True)

print(financials.shape)
```

```python
financials.columns = (
    financials.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("%", "pct")
    .str.replace("&", "_")
    .str.replace("[", "_")
    .str.replace("]", "_")
)
```

```python
financials.rename(columns={'unnamed:_0': 'symbol'}, inplace=True)
```

```python
financials.head()
```

```python
price_cols = [col for col in financials.columns if "price_var" in col]

financials.drop(columns=price_cols, inplace=True)
```

```python
growth_cols = [col for col in financials.columns if "growth" in col and col != "revenue_growth"]

financials.drop(columns=growth_cols, inplace=True)
```

```python
financials.drop(columns=["class"], inplace=True)
```

```python
keep_cols = [
    "symbol",
    "year",
    "revenue",
    "revenue_growth",
    "sector",
    "gross_profit",
    "operating_income",
    "operating_expenses",
    "cost_of_revenue"
]

financials = financials[keep_cols]
```

```python
financials = financials.dropna(subset=['revenue'])
```

```python
financials=financials.fillna(0)
```
